```python
df['TCGA_Classification'] = df['TCGA_Classification'].astype(str)

df['Tissue'] = df['Tissue'].astype(str)
df['Tissue_Sub-type'] = df['Tissue_Sub-type'].astype(str)
```

```python
X = df.drop('IC50', axis=1)  # Replace "target_variable" with the name of your target variable column
y = df["IC50"]
```

```python
obj_feat=list(X.loc[:,X.dtypes=='object'].columns.values)
```

```python
for feature in obj_feat:
    X[feature]=pd.Series(X[feature],dtype="category")
```

```python
lgbmr=lightgbm.LGBMRegressor()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
import time
import numpy as np
```

```python
start=time.time()
lgbmr.fit(X_train,y_train)

y_pred=lgbmr.predict(X_test)

from sklearn.metrics import r2_score as RSquared

lgbm_rsquared=np.sqrt(RSquared(y_test,y_pred))

print("R Squared for LightGBM: ", np.mean(lgbm_rsquared))

end=time.time()
diff=end-start
print('Excecution Time: ',diff)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = lgbmr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```
