```python
import xgboost as xgb
import catboost 
from catboost import CatBoostRegressor, Pool
```

```python
nan_values=df.isna().sum()
nan_values
```

```python
df= df.dropna()
```

```python
nan_values=df.isna().sum()
nan_values
```

```python
df['TCGA_Classification'] = df['TCGA_Classification'].astype(str)

df['Tissue'] = df['Tissue'].astype(str)
df['Tissue_Sub-type'] = df['Tissue_Sub-type'].astype(str)
```

```python
X = df.drop('IC50', axis=1)  # Replace "target_variable" with the name of your target variable column
y = df["IC50"]
```

```python
X
```

```python
y
```

```python
print(df.dtypes)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
pool_train=Pool(X_train,y_train,cat_features=['TCGA_Classification','Tissue','Tissue_Sub-type'])
pool_test=Pool(X_test,cat_features=['TCGA_Classification','Tissue','Tissue_Sub-type'])
```

```python
import time
import numpy as np
```

```python
start=time.time()
cbr=CatBoostRegressor(iterations=100)
```

```python
cbr.fit(pool_train)
y_pred=cbr.predict(X_test)

from sklearn.metrics import r2_score as RSquared

cb_rsquared=np.sqrt(RSquared(y_test,y_pred))

print("R Squared for CatBoost: ", np.mean(cb_rsquared))

end=time.time()
diff=end-start
print('Excecution Time: ',diff)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = cbr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE:", rmse)
```
